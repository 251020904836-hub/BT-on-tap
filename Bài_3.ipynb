```python
mat_khau_dung = "python123"
dem = 0

while dem < 3:
    mk = input("Nhập mật khẩu: ")
    if mk == mat_khau_dung:
        print("Đăng nhập thành công")
        break
    else:
        dem += 1

if dem == 3:
    print("Tài khoản bị khóa")
```
