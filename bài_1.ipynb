```python
gpa = float(input("Nhập GPA (0 - 10): "))
if gpa < 0 or gpa > 10:
    print("lỗi:điểm không hợp lệ")
elif gpa >= 8.5 and gpa <= 10.0 :
    print("Học lực: Xuất sắc")
elif gpa >= 7.0:
    print("Học lực: Giỏi")
elif gpa >= 5.5:
    print("Học lực: Khá")
elif gpa >= 4.0:
    print("Học lực: Trung bình")
else:
    print("Học lực: Yếu")
```
